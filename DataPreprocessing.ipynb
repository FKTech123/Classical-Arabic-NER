```python
# Read the CSV file
df = pd.read_csv('/content/drive/MyDrive/ner/CANERCorpus.csv')

# Initialize variables to keep track of the previous label and its index
prev_label = None
prev_index = -1

# Initialize an empty list to store the IOB tags
iob_tags = []

# Iterate through each row in the DataFrame
for index, row in df.iterrows():
    label = row['Acutal']

    # Convert label to IOB format
    if label == 'O':
        iob_tags.append('O')  # Outside entity
    else:
        if label == prev_label and index == prev_index + 1:
            iob_tags.append(f'I-{label}')  # Inside an entity
        else:
            iob_tags.append(f'B-{label}')  # Beginning of an entity

    # Update previous label and index
    prev_label = label
    prev_index = index

# Add the IOB tags as a new column in the DataFrame
df['IOB'] = iob_tags

# Save the updated DataFrame to a new CSV file
df.to_csv('/content/drive/MyDrive/ner/outputWithIOB.csv', index=False)

print("Conversion completed. IOB tags saved to 'outputWithIOB.csv'.")
```

```python
import csv

# Define the NER tags and their corresponding numeric values
tags = ['o', 'B-Allah', 'I-Allah', 'B-Prophet', 'I-Prophet', 'B-Pers', 'I-Pers', 'B-Loc', 'I-Loc', 'B-Org', 'I-Org', 'B-Meas', 'I-Meas', 'B-Mon', 'I-Mon', 'B-Book',
        'I-Book', 'B-Date', 'I-Date', 'B-Time', 'I-Time', 'B-Rlig', 'I-Rlig', 'B-Sect', 'I-Sect', 'B-Clan', 'I-Clan', 'B-NatOb', 'I-NatOb', 'B-Crime', 'I-Crime',
        'B-Para', 'I-Para', 'B-Hell', 'I-Hell', 'B-Month', 'I-Month', 'B-Day', 'I-Day', 'B-Num', 'I-Num']

ner_tag_map = {tag: i for i, tag in enumerate(tags)}

def add_iob_column_to_csv(input_csv_file, output_csv_file, ner_tag_map):
    with open(input_csv_file, 'r', newline='', encoding='utf-8') as input_file:
        reader = csv.DictReader(input_file)
        fieldnames = reader.fieldnames + ['ner_tag']
        with open(output_csv_file, 'w', newline='', encoding='utf-8') as output_file:
            writer = csv.DictWriter(output_file, fieldnames=fieldnames)
            writer.writeheader()
            for row in reader:
                iob_value = row.get('IOB', '')
                ner_tag = ner_tag_map.get(iob_value)
                if ner_tag is not None:
                    row['ner_tag'] = ner_tag
                else:
                    # Assign a default value or handle the case when 'IOB' value is not found
                    row['ner_tag'] = 0  # Example: Assigning -1 as a default value
                writer.writerow(row)

# Add IOB column to CSV file
add_iob_column_to_csv('/content/drive/MyDrive/ner/outputWithIOB.csv', '/content/drive/MyDrive/ner/outputWithNERtag.csv', ner_tag_map)
```

```python
import csv

class SentenceTokenizer:
    def __init__(self, input_file, output_file, sentence_length=28):
        self.input_file = input_file
        self.output_file = output_file
        self.sentence_length = sentence_length

    def read_csv(self):
        data = []
        with open(self.input_file, 'r', newline='', encoding='utf-8') as csvfile:
            reader = csv.reader(csvfile)
            header = next(reader)
            header.append('Sentence #')
            data.append(header)
            row_count = 0
            sentence_num = 1
            sentence_length_count = 0
            last_tag = None  # Track the last tag in the sentence
            for row in reader:
                row_count += 1
                word, _, iob, _, = row
                if iob == 'B-Pers':  # If current tag is 'B-Pers'
                    if last_tag == 'I-Pers':  # If last tag was 'I-Pers'
                        sentence_num += 1  # Start a new sentence
                last_tag = iob[-4:] if iob != 'O' else None  # Update last tag
                data_row = row[:] + [sentence_num]
                data.append(data_row)
                sentence_length_count += 1
                if sentence_length_count == self.sentence_length:
                    sentence_length_count = 0
                    sentence_num += 1
            # Ensure the last row of the sentence ends with 'O'
            if last_tag != 'O':
                data_row = ['O'] * (len(header) - 1) + [sentence_num]
                data.append(data_row)
        return data

    def write_csv(self, data):
        with open(self.output_file, 'w', newline='', encoding='utf-8') as csvfile:
            writer = csv.writer(csvfile)
            for row in data:
                writer.writerow(row)

    def tokenize_sentences(self):
        data = self.read_csv()
        self.write_csv(data)

# # Usage Example:
# input_file = 'outputWithNERtag.csv'
# output_file = 'corpus.csv'
# tokenizer = SentenceTokenizer(input_file, output_file)
# tokenizer.tokenize_sentences()

# Usage Example:
input_file = '/content/drive/MyDrive/ner/outputWithNERtag.csv'
output_file = '/content/drive/MyDrive/ner/corpus.csv'
tokenizer = SentenceTokenizer(input_file, output_file)
tokenizer.tokenize_sentences()
```

Now, we have to ask ourself the question: what is a training example in the case of NER, which is provided in a single forward pass? A training example is typically a sentence, with corresponding IOB tags. Let's group the words and corresponding tags by sentence:

```python
# pandas has a very handy "forward fill" function to fill missing values based on the last upper non-nan value
data = data.fillna(method='ffill')
data.head()
```

```python
# let's create a new column called "sentence" which groups the words by sentence
# data['sentence'] = data[['Sentence #','Word','IOB']].groupby(['Sentence #'])['Word'].transform(lambda x: ' '.join(x))
data['sentence'] = data[['Sentence #','Word','IOB']].groupby(['Sentence #'])['Word'].transform(lambda x: ' '.join(str(word) for word in x))
# let's also create a new column called "word_labels" which groups the tags by sentence
data['word_labels'] = data[['Sentence #','Word','IOB']].groupby(['Sentence #'])['IOB'].transform(lambda x: ','.join(x))
data.head()
```
